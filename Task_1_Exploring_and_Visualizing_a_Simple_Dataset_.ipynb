# Task 1 Exploring and Visualizing a Simple Dataset

```python
#import libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
df = pd.read_csv('iris.csv')
```

# EDA

```python
# Show a dataset
df.head()
```

```python
#display  total row and column
print("shape:", df.shape)
```

```python
# Display a columns name
df.columns
```

# summary statistics.

```python
# Display dataset structure and data types
df.info()
```

```python
# Display Summary statistics including mean, min, max, and quartiles
df.describe()
```

# Visualize the dataset

```python
# Scatter plot to show relationships between features.
plt.scatter(df['sepal.length'],df['sepal.width'])
plt.xlabel('sepal.length')
plt.ylabel('sepal.width')
plt.title(' Relationship between sepal.length and sepal.width ')
plt.show()
```

```python
# Histogram plot for a One Feature
sns.histplot(df['sepal.length'])
plt.title('Distribution of sepal.length')
plt.show()

# Histogram plot for a Multiple Feature
df.hist(figsize=(8,4))
plt.tight_layout()
plt.show()
```
